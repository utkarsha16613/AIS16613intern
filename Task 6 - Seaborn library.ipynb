* # Seaborn library
  Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies the process of creating complex visualizations and is particularly useful for visualizing datasets in data science and machine learning projects.

* Basic Usage  
Here's an example of how to create a simple plot using Seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a simple scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")

# Show the plot
plt.show()
```

* # 1.Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")

# Show the plot
plt.show()
```

* # 2.Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")

# Show the plot
plt.show()
```

* # 3.Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
titanic = sns.load_dataset("titanic")

# Create a bar plot
sns.barplot(data=titanic, x="class", y="survived", hue="sex")

# Show the plot
plt.show()
```

* # 4.Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a histogram
sns.histplot(data=tips, x="total_bill", bins=20)

# Show the plot
plt.show()
```

* # 5.Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a box plot
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")

# Show the plot
plt.show()
```

* # 6.Pair Plot  
Pair plots are useful for visualizing pairwise relationships in a dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
iris = sns.load_dataset("iris")

# Create a pair plot
sns.pairplot(iris, hue="species")

# Show the plot
plt.show()
```

* # 7.Regression plot
 Regression plots in Seaborn are used to visualize the relationship between two variables along with a fitted regression line. Seaborn provides several functions to create regression plots, such as sns.regplot(), sns.lmplot(), and sns.residplot().

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a regression plot
sns.regplot(data=tips, x="total_bill", y="tip")

# Show the plot
plt.show()
```

* # 8.Residual Plot
 A residual plot is used to visualize the residuals of a regression model. Residuals are the differences between observed and predicted values. Seaborn's residplot function is specifically designed for creating residual plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a residual plot
sns.residplot(data=tips, x="total_bill", y="tip")

# Show the plot
plt.show()
```

* # 9.Plotting a Distplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([0, 1, 2, 3, 4, 5])

plt.show()
```

* # 10.Plotting a Distplot Without the Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([0, 1, 2, 3, 4, 5], hist=False)

plt.show()
```
